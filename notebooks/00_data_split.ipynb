```python
import json
import os
import random
```

```python
PATH = f'../data/processed'
DEST = f'../data/processed/splits'
```

```python
examples = []
for filename in os.listdir(PATH):
    if filename.endswith(".jsonl"):  
        file_path = os.path.join(PATH, filename)
        
        try:
            with open(file_path, "r", encoding="utf-8") as f:
                examples.extend([json.loads(line) for line in f])
            
            print(filename, len(examples))
        
        except UnicodeDecodeError:
            print(f"Error decoding {filename} with utf-8 encoding. Trying latin-1...")
            with open(file_path, "r", encoding="latin-1") as f:
                examples.extend([json.loads(line) for line in f])
            
            print(filename, len(examples))
        
        except Exception as e:
            print(f"An error occurred with {filename}: {e}")
```

```python
random.shuffle(examples)
n = len(examples)
```

```python
train = examples[:int(n * 0.8)]   # 8,000
val   = examples[int(n * 0.8):int(n * 0.9)]  # 1,000
test  = examples[int(n * 0.9):]   # 1,000
```

```python
os.makedirs(DEST, exist_ok=True)  # make sure the folder exists

for split, name in [(train, "train"), (val, "val"), (test, "test")]:
    with open(os.path.join(DEST, f"{name}.jsonl"), "w", encoding="utf-8") as f:
        for ex in split:
            f.write(json.dumps(ex, ensure_ascii=False) + "\n")
```

```python
print(f"Train: {len(train)}, Val: {len(val)}, Test: {len(test)}")
```
