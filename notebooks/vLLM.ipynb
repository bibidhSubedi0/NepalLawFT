```python
from vllm import LLM, SamplingParams

llm = LLM(
    model="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    dtype="float16"
)

sampling_params = SamplingParams(
    temperature=0.7,
    max_tokens=200
)

outputs = llm.generate(
    ["Explain what a black hole is in simple terms."],
    sampling_params
)

print(outputs[0].outputs[0].text)
```

```python
import time

questions = [
    "What is a black hole?",
    "Explain quantum physics simply.",
    "What is the theory of relativity?",
    "How does photosynthesis work?",
    "What is DNA?",
]

# BATCHED — all 5 at once (vLLM style)
start = time.time()
outputs = llm.generate(questions, sampling_params)
batched_time = time.time() - start

print(f"BATCHED — all 5 questions simultaneously: {batched_time:.2f}s\n")
for output in outputs:
    print(f"Q: {output.prompt}")
    print(f"A: {output.outputs[0].text[:100]}...")
    print("─" * 50)
```

```python
# SEQUENTIAL — one at a time (naive model.generate() style)
start = time.time()
for q in questions:
    out = llm.generate([q], sampling_params)
sequential_time = time.time() - start

print(f"\nSEQUENTIAL — one at a time: {sequential_time:.2f}s")
print(f"Batched was {sequential_time/batched_time:.1f}x faster")
```
